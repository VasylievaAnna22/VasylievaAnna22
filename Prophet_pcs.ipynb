## Loading and Preparing the Dataset
Load the dataset and perform necessary preprocessing.

```python
# Load dataset
df = pd.read_csv('dataset_pcs.csv')

# Convert 'period' to datetime
df['period'] = pd.to_datetime(df['period'])

# Select only required columns
df = df[['period', 'sku_id', 'qty_total', 'avg_discount_perc_by_goods', 'oos__by_goods', 'sin_month', 'cos_month']]
```

Combine all forecasts and save to a CSV file.

```python
forecast_all = []

for sku in df['sku_id'].unique():
    sku_df = df[df['sku_id'] == sku][['period', 'qty_total']].rename(columns={'period': 'ds', 'qty_total': 'y'})
    if sku_df.shape[0] < 10 or sku_df['y'].nunique() <= 1:
        continue

    model = Prophet()
    model.fit(sku_df)
    future = model.make_future_dataframe(periods=28)
    forecast = model.predict(future)
    forecast['sku_id'] = sku
    forecast_all.append(forecast[['ds', 'yhat', 'sku_id']])
    final_forecast_df = pd.concat(forecast_all, ignore_index=True)
final_forecast_df.to_csv('forecast.csv', index=False)
```
